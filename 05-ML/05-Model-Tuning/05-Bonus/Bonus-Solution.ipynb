```python
import pandas as pd

data = pd.read_csv("data.csv")

data.head()
```

The dataset describes tumors that are either malignant or benign.

ðŸ‘‡ Using what you have learnt this week, build a model to flag as many malginant tumors as possible. Have fun!

## Scale

```python
from sklearn.preprocessing import MinMaxScaler

# Select only the features 
X = data.drop(columns="target")

# Fit scaler to features
scaler = MinMaxScaler().fit(X)

# Put transformed data back to dataframe
data.loc[:,'mean radius':'worst fractal dimension'] = scaler.transform(X)

data.head()
```

## Encode label

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

encoder.fit(data.target)

data["target"] = encoder.transform(data.target)

data.head()
```

## Holdout Split

```python
from sklearn.model_selection import train_test_split

data_train, data_test = train_test_split(data, test_size = 0.3)
```

## SVM Randomized Search

```python
from sklearn.svm import SVC
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import uniform

# Prepare data
X_train = data_train.drop(columns="target")
y_train = data_train['target']

# Instanciate model
model = SVC()

# Hyperparameter search space
search_space = {'kernel' : ['linear', 'poly', 'rbf', 'sigmoid'], 'C': uniform(0.1,20)}

#scoring
scoring=['accuracy','precision','recall']

# Instanciate Random Search
search = RandomizedSearchCV(model,
                            param_distributions = search_space,
                            n_jobs=-1, scoring = scoring,
                            refit='recall', # Specify best estimator based on recall score
                            cv = 4, 
                            n_iter = 20)

# Fit data to Grid Search
search.fit(X_train, y_train)
```

```python
search.best_params_
```

```python
# Full results in a dataframe
full_results = pd.DataFrame(search.cv_results_)

full_results.head()
```

## Test Generalisation

```python
X_test = data_test.drop(columns = 'target')
y_test = data_test['target']

# Extract best model
model = search.best_estimator_

#this gives the accuracy
model.score(X_test,y_test)
```

## Classification Report

```python
from sklearn.metrics import classification_report

# Use model to predict
y_pred = model.predict(X_test)

# Compare predictions and true values to produce classification report
print(classification_report(y_test,y_pred))
```

## Confusion matrix

```python
from sklearn.metrics import plot_confusion_matrix

plot_confusion_matrix(model, X_test, y_test)
```
