```python
import pandas as pd

data = pd.read_csv("data/creditcard.csv")

data.head()
```

```python
df_class_0 = data[data['Class'] == 0]
df_class_1 = data[data['Class'] == 1]
```

```python
df_class_0_sample = df_class_0.sample(frac=0.3)
```

```python
df_class_1_sample = df_class_1.sample(len(df_class_0_sample), replace = True)
```

```python
data_balanced = pd.concat([df_class_0_sample, df_class_1_sample], axis=0).sample(frac=1)
```

```python
data_balanced
```

```python
target = 'Class'
features = list(data_balanced.columns)
features.remove(target)

X = data_balanced[features]
y = data_balanced[target]
```

```python
from sklearn.metrics import recall_score
from sklearn.model_selection import cross_validate
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import precision_recall_curve
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=2000)
```

```python
data_balanced['prob_ok'], data_balanced['prob_fraud'] = cross_val_predict(model,
                                                   X,
                                                   y,
                                                   method = "predict_proba").T

precision, recall, thresold = precision_recall_curve(y, data_balanced['prob_fraud'])
```

```python
scores = pd.DataFrame({"thresold":thresold,
                       "precision":precision[:-1],
                       "recall":recall[:-1]})

scores
```

```python
new_threshold = scores[scores['recall'] >= 0.9].thresold.max()
new_threshold
```
