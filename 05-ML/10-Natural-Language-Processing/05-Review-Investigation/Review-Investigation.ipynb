```python
import pandas as pd
import numpy as np

df = pd.read_csv("https://wagon-public-datasets.s3.amazonaws.com/Machine%20Learning%20Datasets/reviews.csv")

df.head()
```

👇 Analyse the comment reviews to understand what causes the bad review scores.

```python
new_df = df[['review_score','review_comment_title','review_comment_message']]
new_df = new_df[new_df['review_comment_message'].notna()]
new_df = new_df[new_df['review_score'].notna()]
```

```python
new_df['review_comment_title'] = new_df['review_comment_title'].replace(np.nan, "")
new_df['review_comment_title'] = new_df['review_comment_title'].astype('string')
new_df['review_comment_message'] = new_df['review_comment_message'].astype('string')
```

```python
new_df['review'] = new_df['review_comment_title'] + " "+ new_df['review_comment_message']
new_df = new_df[['review_score',"review"]]
```

```python
new_df
```

```python
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize

stop_words = set(stopwords.words('portuguese')) 
lemmatizer = WordNetLemmatizer()
```

```python
import string 

def clean(s):
    s_copy = s.copy()
    for punct in string.punctuation:   #remove punctuation
        s_copy = [text.replace(punct, '') for text in s_copy]
    s_copy = [text.lower() for text in s_copy]  #lowercase
    s_copy = [''.join(word for word in text if not word.isdigit()) for text in s_copy] #delete numbers
    s_copy = [[w for w in word_tokenize(text) if not w in stop_words] for text in s_copy] #tokenize
    s_copy = [" ".join([lemmatizer.lemmatize(word) for word in text]) for text in s_copy] #lemmatize
    return s_copy

clean(new_df['review'])
```

```python
new_df['review'] = clean(new_df['review'])
```

```python
len(new_df)
```

```python
new_df = new_df[new_df['review_score'].isin(['5', '4', '2', '1', '3',1, 3, 4, 5, 2])]
```

```python
new_df.review_score.unique()
```

```python
new_df['review_score'] = new_df['review_score'].astype('int')
bad_reviews = new_df[new_df['review_score'].isin([1,2,3])]
```

```python
bad_reviews['review'] = [text.replace('produto','') for text in bad_reviews['review']]
```

```python
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(ngram_range=(1,2)).fit(bad_reviews.review)

data_vectorized = vectorizer.transform(bad_reviews.review)

lda_model = LatentDirichletAllocation(n_components=2).fit(data_vectorized)

def print_topics(model, vectorizer):
    for idx, topic in enumerate(model.components_):
        print("Topic %d:" % (idx))
        print([(vectorizer.get_feature_names()[i], topic[i])
                        for i in topic.argsort()[:-10 - 1:-1]])
        

print_topics(lda_model, vectorizer)
```

```python
from deep_translator import GoogleTranslator

def print_topics(model, vectorizer):
    for idx, topic in enumerate(model.components_):
        print("Topic %d:" % (idx))
        print([(GoogleTranslator(source='auto', target='fr').translate(vectorizer.get_feature_names()[i]), topic[i])
                        for i in topic.argsort()[:-10 - 1:-1]])
        

print_topics(lda_model, vectorizer)
```
