# Subploting

Sometimes we need more than one plot, we might want to compare 3 similar graph or show an evolution.
To do so we can create subplots in our `figure` object to create a **grid** of plot.

<img src="https://matplotlib.org/3.1.1/_images/sphx_glr_axes_margins_001.png" alt="subplot" width=400>

### Side by side plot

Change the following code to show the two barplots **side by side**.
To do this use the [plt.subplot()](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.subplot.html?highlight=subplot#matplotlib.pyplot.subplot) function.

**hint :** you will need to use *axes*.

```python
import matplotlib.pyplot as plt
import numpy as np

wallet_2015 = {"currency": ["bitcoin","ethereum","euro"], "value_in_euro": [234,74,530]}
wallet_2020 = {"currency": ["bitcoin","ethereum","euro"], "value_in_euro": [234*4,74*7,530*1.3]}

plt.bar(wallet_2015["currency"], wallet_2015["value_in_euro"])
plt.show()
plt.bar(wallet_2020["currency"], wallet_2020["value_in_euro"])
plt.show()
```

### Custom layout

In the following code we have a third graph that we want to print along side the two barplot.

In this layout the **two barplots** should be on **top** and the **line plot** at the **bottom**.

Change the folling code to get this layout.

You will need [plt.figure()](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.figure.html#matplotlib.pyplot.figure) and [figure.add_gridspec()](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.figure.Figure.html#matplotlib.figure.Figure.add_gridspec) to create it.

```python
bitcoin = [3,3,4,5,5,6,11,12,13,14,16,14,18,15,19,18,19,15,10,8,9,9,9,12,13]
ethereum = [1,1,1,2,2,3,3,4,4,4,4,8,9,12,10,9,14,15,10,8,6,5,5,8,7]
euro = [7,7,7,7.6,7,7.8,7,7,7,7,7.6,7,7,7,7,6.8,6.9,7,7,7,7,7,7.2,7.2,7.3]

plt.plot(bitcoin)
plt.plot(ethereum)
plt.plot(euro)
plt.show()
plt.bar(wallet_2015["currency"], wallet_2015["value_in_euro"])
plt.show()
plt.bar(wallet_2020["currency"], wallet_2020["value_in_euro"])
plt.show()
```
